1. Import libraries

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(tf.__version__)
```

2. Load datasets

```python
dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
dfeval = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv')
y_train = dftrain.pop('survived')
y_eval = dfeval.pop('survived')
```

3. Quick data exploration

```python
print(dftrain.head())
print(dftrain.describe())
print(dftrain.shape)
```

4. Feature columns

```python
CATEGORICAL_COLUMNS = ['sex', 'n_siblings_spouses', 'parch', 'class', 'deck', 'embark_town', 'alone']
NUMERIC_COLUMNS = ['age', 'fare']

feature_columns = []
for feature in CATEGORICAL_COLUMNS:
    vocab = dftrain[feature].unique()
    feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature, vocab))

for feature in NUMERIC_COLUMNS:
    feature_columns.append(tf.feature_column.numeric_column(feature, dtype=tf.float32))
```

5. Input functions

```python
def make_input_fn(data_df, label_df, num_epochs=10, shuffle=True, batch_size=32):
    def input_function():
        ds = tf.data.Dataset.from_tensor_slices((dict(data_df), label_df))
        if shuffle:
            ds = ds.shuffle(buffer_size=len(data_df))
        ds = ds.batch(batch_size).repeat(num_epochs)
        return ds
    return input_function

train_input_fn = make_input_fn(dftrain, y_train)
eval_input_fn = make_input_fn(dfeval, y_eval, num_epochs=1, shuffle=False)
```

6. Create a DNNClassifier (More advanced than LinearClassifier from the previous models)

```python
linear_est = tf.estimator.LinearClassifier(
    feature_columns=feature_columns,
    optimizer=tf.optimizers.Ftrl(learning_rate=0.1),  # Tweakable optimizer
    n_classes=2
)
```

7. Train and evaluate the model

```python
linear_est.train(train_input_fn)
result = linear_est.evaluate(eval_input_fn)

print("\nModel Evaluation:")
for key, value in result.items():
    print(f"{key}: {value}")
```

8. Predict survival probabilities

```python
pred_dicts = list(linear_est.predict(eval_input_fn))
probs = pd.Series([pred['probabilities'][1] for pred in pred_dicts])
```

Plot survival probability distribution

```python
probs.plot(kind='hist', bins=20, color='green', edgecolor='black')
plt.title("Predicted Survival Probability Distribution")
plt.xlabel("Probability of Survival")
plt.ylabel("Frequency")
plt.show()
```

Compare predicted probabilities vs. actual labels

```python
plt.scatter(y_eval, probs, alpha=0.6)
plt.xlabel("Actual Survival (0=No, 1=Yes)")
plt.ylabel("Predicted Probability")
plt.title("Predicted vs Actual Survival")
plt.grid(True)
plt.show()
```
